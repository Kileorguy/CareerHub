```python
df = pd.read_csv('./final_data.csv')
df.head(5)
```

```python
import pandas as pd
import uuid

all_jobs = df[['search_position', 'company', 'job_level', 'job_summary', 'job_level', 'job_skills', 'search_country', 'search_city', 'id']]

skills_columns = ["skill_id", "skill_name"]
registered_skills = {}
skills_df = pd.DataFrame(columns=skills_columns)

jobs_columns = ["job_id", "job_name", "job_level", "job_summary", "company_id", "company", "search_country", "search_city"]
jobs_df = pd.DataFrame(columns=jobs_columns)

jobs_and_skills_merged_columns = ["job_id", "skill_id"]
jobs_and_skills_merged_df = pd.DataFrame(columns=jobs_and_skills_merged_columns)

for index, row in all_jobs.iterrows():
    if isinstance(row['job_skills'], str):
        job_id = str(uuid.uuid4())
        new_job = pd.DataFrame([{
            "job_id": job_id, 
            "job_name": row['search_position'].lower(),
            "job_summary": row['job_summary'],
            "job_level": row['job_level'],

            "company_id": row['id'],
            "company": row['company'],
            "search_country": row['search_country'],
            "search_city": row['search_city'],
        }])
        jobs_df = pd.concat([jobs_df, new_job], ignore_index=True)
        for skill in row['job_skills'].split(', '):
            if skill.lower() not in registered_skills:
                skill_id = str(uuid.uuid4())
                new_skill = pd.DataFrame([{
                    "skill_id": skill_id, 
                    "skill_name": skill.lower()
                }])
                skills_df = pd.concat([skills_df, new_skill], ignore_index=True)
                registered_skills[skill.lower()] = skill_id
            new_job_skill = pd.DataFrame([{
                "job_id": job_id,
                "skill_id": registered_skills[skill.lower()],
            }])
            jobs_and_skills_merged_df = pd.concat([jobs_and_skills_merged_df, new_job_skill])

skills_df.to_csv('skills.csv', index=False)
jobs_df.to_csv('jobs.csv', index=False)
jobs_and_skills_merged_df = jobs_and_skills_merged_df.drop_duplicates()
jobs_and_skills_merged_df.to_csv('jobs_skills_map.csv', index=False)
```
