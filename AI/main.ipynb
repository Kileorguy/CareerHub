```python
import pandas as pd
import numpy as np
import uuid
```

```python
root_dataset = './Dataset'
job_skills_df = pd.read_csv(f"{root_dataset}/job_skills.csv",)
print('1')
job_summary_df = pd.read_csv(f"{root_dataset}/job_summary.csv")
print('2')
linked_in_df = pd.read_csv(f"{root_dataset}/linkedin_job_postings.csv")
print('3')
```

```python
main_linked_in = linked_in_df.copy()[:250]
print(len(main_linked_in))
merged_df = pd.merge(main_linked_in,job_summary_df,left_on='job_link',right_on='job_link')
new_merged_df = pd.merge(merged_df,job_skills_df,left_on='job_link',right_on='job_link')
```

```python
print(len(new_merged_df))
display(new_merged_df.head(3))
```

```python
new_merged_df['id'] = new_merged_df.apply(lambda _: uuid.uuid4(), axis=1)

new_merged_df.to_csv('final_data.csv')
```

```python
df = pd.read_csv('./final_data.csv')

skill_set = set()

all_skills = df['job_skills']
for skills in all_skills:
    if isinstance(skills, str):
        for skill in skills.split(', '):
            skill_set.add(skill)

new_df = pd.DataFrame({"job_skills": list(skill_set)})
new_df.to_csv('job_skills.csv', index=False)
```
